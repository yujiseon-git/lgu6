##3. for 문의 활용: 음식 주문 프로세스 구현

```python
menus=['된장찌개', '삼겹살', '설렁탕', '냉면', '칼국수', '김치찌개']
requests=['된장찌개', '김치찌개', '해물찜']

for item in requests:
  if item in menus:
    print(f"{item} 주문합니다.")
  else:
    print(f"\n죄송합니다. 저희는 {item}이 없습니다.")
```

```python
ready=[]
ready.append('된장찌개')
```

```python
menus=['된장찌개', '삼겹살', '설렁탕', '냉면', '칼국수', '김치찌개']
requests=['된장찌개', '김치찌개', '해물찜']
ready=[]

for item in requests:
  if item in menus:
    print(f"{item} 주문합니다.")
    ready.append(item)
    print(f"{ready} 준비해주세요.")
  else:
    print(f"\n죄송합니다. 저희는 {item}이 없습니다.")
print(f"\n이제 {ready} 음식을 준비하겠습니다.")
```

## 5. while 문의 사용

```python
i=0
while i<5:
  print (i)
  i+=1 # i=i+1과 같은 의미
```

```python
i=0
total=0
while i<5:
  i=i+1
  total+=i
  print(i, total) # 반복문 안이므로 i와 total 변수의 결과를 반복 출력
print(i, total) # 반복문 밖이므로 i와 total 변수의 최종 결과만 출력
```

```python
i=0
total=0
while i<100:
  print(i)
  i+=10
  total+=i
print(i, total)
```

```python
i=0
total=0
for i in range(0, 101, 10):
  print(i)
  total+=i
print(i, total)
```

```python
i=0
weather=['rain', 'snow', 'good']
while i<10:
  print(i)
  if weather[i%3]=='rain':
    print('비가 오니 등산 안 한다.')
  elif weather[i%3]=='snow':
    print('눈이 오니 등산 안 한다.')
  else:
    print("%s 날씨이니 등산한다."%weather[i%3])
    print(f"{weather[i%3]} 날씨이니 등산한다.")
  i=i+1
```
