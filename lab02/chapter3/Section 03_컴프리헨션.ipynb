#Section 03. 컴프리헨션

```python
# 일반 리스트 생성
even_numbers=[0, 2, 4, 6, 8]
even_numbers
```

```python
# 컴프리헨션을 이용한 리스트 생성
squares=[x * x for x in range(5)]
squares
```

```python
# 짝수의 제곱을 원소로 가지는 리스트 컴프리헨션
even_squares=[x * x for x in even_numbers]
even_squares
```

```python
# 주어진 크기의 원소를 가지는 리스트 컴프리헨션
zeros=[0 for _ in even_numbers]
zeros
```

```python
# 0에서 9까지의 짝수를 원소로 가지는 리스트 컴프리헨션
even_numbers=[x for x in range(10) if x % 2==0]
even_numbers
```

```python
# 구구단 리스트 컴프리헨션 1
pairs=[i*j for i in range(2, 10) for j in range(1, 10)]
print(pairs)
```

```python
# 2단 리스트 컴프리헨션
googoo=[]
for i in range(2, 10) :
  for j in range(1, 10) :
    googoo.append(i * j)
googoo=[i*j for i in range(2, 10) for j in range(1, 10)]
googoo[:9]
```

```python
# 구구단 리스트 컴프리헨션 2
for i in range(2, 10) :
  for j in range(1, 10) :
    print("{} * {}={}".format(i, j, googoo[(j-1)+((i-2)*9)]))
```

```python
# 구구단 리스트 컴프리헨션 3
increasing_pairs=[(x, y)
  for x in range(10)
  for y in range(x+1, 10)]
increasing_pairs
googoo=[(x, y) for x in range(2, 10) for y in range(1, 10)]
googoo
```

```python
# 딕셔너리 컴프리헨션
square_dict={i: i * i for i in range(5)}
square_dict
```

```python
# 집합 컴프리헨션
square_set={i * i for i in [1, -1, 2]}
square_set
```
