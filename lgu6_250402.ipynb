# loc vs iloc
- 코드 비교

## loc

```python
#tips.loc[tips['day'] =='Sat']
#tips.loc[행, 열]
tips.loc[0:1, ['total_bill', 'tip', 'day']]
```

```python
#iloc
tips.iloc[0:1, [0, 1, 4]]
```

```python
# 가져오고 싶은 칼럼
tips.loc[행 조건식만 만들자!, :]
```

```python
#total_bill 11이하인것만 조회
#tips.loc[;, ;]
tips.loc[tips['total_bill'] < 11, :].reset_index(drop=True)
```

```python
tips.loc[(tips['time'] == 'Dinner') & (tips['total_bill'] <= 11), :]
# 조건식 2개
# tips.loc[(조건식) & (조건식), :]
```

```python
#time Dinner이면서 total_bill이 11이하인 것만 조회
result = tips.loc[tips['time']=='Dinner', :]
result.loc[result['total_bill'] <= 11, :]
```

```python
tips.loc[tips['day'] =='Sat', ['total_bill', 'tip', 'day']].reset_index(drop=True)
```
