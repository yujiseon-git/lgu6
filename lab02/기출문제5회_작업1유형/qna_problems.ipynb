## 문제 1. 
- 다음 데이터에서 conventional 컬럼의 특수문자를 제거하고, IQR(3분위수-1분위수)를 구하시오. 
- 소수점 첫째 자리에서 반올림

```python
# 코드
data = pd.read_csv('5_1_1.csv')
data.head()
```

```python
# 영어, 숫자, 한국어를 제외한 모든 문자 제거 
# 코드 
data.loc[:,'conventional'] = data.loc[:,'conventional'].str.replace(r'[^a-zA-Z0-9가-힣]', '',regex=True)
data.head(5)
```

```python
# 코드
q3 = data['conventional'].quantile(0.75)
q1 = data['conventional'].quantile(0.25)
iqr = q3 - q1
iqr_rounded = round(iqr, 1)

print('IQR (반올림):', iqr_rounded)
```

## 문제 2. 
- 위 데이터에서 흑인 또는 히스패닉 비율(minority) / 빈곤율(poverty) > 2이며, 도시 유형(city)이 state인 도시의 범죄율(crime) 평균을 구하시오. 
- 소수점 첫째 자리에서 반올림

```python
# 코드
data_filtered = data[(data['minority'] / data['poverty']) > 2] 
data_filtered = data_filtered[data_filtered['city'] == 'state'] 
crime_mean_rounded = data_filtered['crime'].mean() 
crime_mean_rounded = round(crime_mean_rounded ,1) 
print('범죄율 평균:',crime_mean_rounded) 
```
