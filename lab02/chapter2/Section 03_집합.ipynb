#Section 03. 집합

## 1. 집합 생성과 자료 저장

```python
prices_list=[85000, 140000, 200000] # 리스트
prices_set=set(prices_list) # 리스트 변수 앞에 set 함수 사용
prices_set
```

```python
prices_tuple=(85000, 140000, 200000) # 튜플
prices_set=set(prices_tuple) # 튜플 변수 앞에 set 함수 사용
prices_set
```

## 2. 집합의 중복 데이터 제거

```python
prices_set={85000, 140000, 200000}
prices_set.add(5555)
prices_set
```

```python
# 리스트는 원소 중복 가능
prices_list=[85000, 140000, 200000]
prices_list.append(85000)
prices_list
```

```python
# 집합은 원소 중복 불가능
prices_set=set(prices_list)
len(prices_set)
prices_set
```
