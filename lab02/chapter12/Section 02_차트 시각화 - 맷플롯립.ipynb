##1. 선형 차트
### 기본 선형 차트

```python
from matplotlib import pyplot as plt
years=[1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
gdp=[300, 543, 1075, 2862, 5979, 10289, 14958, 18320]
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title('GDP')
plt.ylabel('$')
plt.show()
plt.savefig('viz_gdp.png')
```

###옵션 설정

### 계열별 선형 차트

```python
dataset=[[25, 100], [52, 256], [38, 152], [32, 140], [25, 150], [45, 183], [40, 175], [55, 203],
[28, 152], [42, 198]]
actual=[j[1] for j in dataset]
predicted=[115, 242, 176, 148, 115, 209, 185, 256, 129, 195]
error=[(x - y) for x, y in zip(actual, predicted)]
xs=[i for i, _ in enumerate(error)]
plt.plot(xs, actual, 'g-', label='x') # 녹색 실선
plt.plot(xs, predicted, 'r-.', label='y') # 빨간색 점선
plt.plot(xs, error, 'b', label='error') # 파란색 점선
plt.legend(loc=9) # 범례 표시: loc 숫자로 위치 설정
plt.xlabel('mode')
plt.title('actual vs predicted')
plt.show()
```

### 그리드 표시

```python
import matplotlib.pyplot as plt
import numpy as np
dataset=[[25, 100], [52, 256], [38, 152], [32, 140], [25, 150], [45, 183], [40, 175], [55, 203],
[28, 152], [42, 198]]
apt=[i[0] for i in dataset]
kw=[i[-1] for i in dataset]
xpoints=np.array(apt)
ypoints=np.array(kw)
plt.xlabel('apt')
plt.ylabel('kw')
plt.title('apt vs kw')
plt.scatter(xpoints, ypoints)
plt.grid()
plt.show()
```

### 구간 확대 및 축소

```python
import matplotlib.pyplot as plt
import numpy as np
dataset=[[25, 100], [52, 256], [38, 152], [32, 140], [25, 150], [45, 183], [40, 175], [55, 203],
[28, 152], [42, 198]]
apt=[i[0] for i in dataset]
kw=[i[-1] for i in dataset]
xpoints=np.array(apt)
ypoints=np.array(kw)
plt.xlabel('apt')
plt.ylabel('kw')
plt.title('apt vs kw')
plt.scatter(xpoints, ypoints)
plt.xlim(10, 70) # x축은 10부터 70까지 표시
plt.ylim(0, 300) # y축은 0부터 300까지 표시
plt.grid()
plt.show()
```

### 하위 차트 표시

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1, 10)
y1=x*5
y2=x*1
y3=x*0.3
y4=x*0.2
# 왼쪽 위에 표시
plt.subplot(2, 2, 1)
plt.plot(x, y1)
# 오른쪽 위에 표시
plt.subplot(2, 2, 2)
plt.plot(x, y2)
# 왼쪽 아래 표시
plt.subplot(2, 2, 3)
plt.plot(x, y3)
# 오른쪽 아래 표시
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1, 10)
y1=x*5
y2=x*1
y3=x*0.3
y4=x*0.2
plt.figure(figsize=(20,5))
# 왼쪽 위에 표시
plt.subplot(2, 2, 1)
plt.plot(x, y1)
# 오른쪽 위에 표시
plt.subplot(2, 2, 2)
plt.plot(x, y2)
# 왼쪽 아래 표시
plt.subplot(2, 2, 3)
plt.plot(x, y3)
# 오른쪽 아래 표시
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.show()
```

##2. 막대형 차트, 히스토그램, 산포도, 원형 차트
### 막대형 차트

```python
from matplotlib import pyplot as plt
movies=['cinema', 'avengers', 'squid game', '007', 'iron man']
score=[5, 10, 3, 8, 9]
plt.bar(movies, score)
plt.ylabel('score')
plt.title('Movies')
plt.xticks([i+0.1 for i, _ in enumerate(movies)], movies) # 0.1에서 막대형 차트 그리기 시작
plt.show()
```

```python
plt.barh(movies,score)
plt.ylabel('Movies')
plt.title('score')
plt.yticks([i + 0.1 for i, _ in enumerate(movies)], movies) # 0.1부터 y축 그리기 시작
plt.show()
```

```python
plt.bar(movies, score, color='red', width=0.3)
plt.ylabel('score')
plt.title('Movies')
plt.xticks([i + 0.1 for i, _ in enumerate(movies)], movies)
plt.show()
```

### 히스토그램

```python
from collections import Counter
grades=[83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
histogram=Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x+5 for x in histogram.keys()], # 오른쪽으로 5만큼 이동해서 그리기 시작
histogram.values(),
10,
edgecolor=(0, 0, 0))
plt.axis([-5, 105, 0, 5]) # x축 범위와 y축 범위 지정
plt.xticks([10 * i for i in range(11)]) # x축 라벨 표시
plt.xlabel('Decile')
plt.ylabel('# of Students')
plt.title('Distribution of Exam 1 Grades')
plt.savefig('viz_grades.png') # 그림 파일 저장
```

### 산포도

```python
from matplotlib import pyplot as plt
height=[175, 170, 162, 171, 186, 175, 166, 169, 189, 178] # x 독립변수
weight=[ 70, 65, 61, 63, 78, 64, 60, 64, 88, 76] # y 종속변수
labels=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'I', 'j']
plt.scatter(height, weight)
```

```python
from matplotlib import pyplot as plt
height=[175, 170, 162, 171, 186, 175, 166, 169, 189, 178] # x 독립변수
weight=[70, 65, 61, 63, 78, 64, 60, 64, 88, 76] # y 종속변수
labels=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'I', 'j']
plt.scatter(height, weight)
for i, j, k in zip(labels, height, weight):
  plt.annotate(i,
               xy=(j, k), # 레이블 표시
               xytext=(5, -5), # 텍스트 위치
  textcoords='offset points')
plt.title('height vs. weight')
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```
